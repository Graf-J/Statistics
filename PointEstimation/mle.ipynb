```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

$$\mu = \frac{1}{n} \cdot \sum_{i = 0}^{n} x_i$$

```python
def mean(x: np.ndarray) -> float:
    return (1 / len(x)) * x.sum()
```

$$\sigma^{2}= \frac{1}{n} \cdot \sum_{i = 0}^{n} (x_i - \mu)^{2}$$

```python
def variance(x: np.ndarray, mu: float) -> float:
    return (1 / len(x)) * np.pow(x - mu, 2).sum()
```

$$f(x \mid \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

```python
def plot_normal_distribution(x: np.ndarray) -> None:
    mu = mean(x)
    var = variance(x, mu)
    std = np.sqrt(var)

    x_distribution = np.linspace(-10, 12, 500)
    normal_distribution = lambda x, mu, std: (1 / np.sqrt(2 * np.pi * std)) * np.exp(-np.pow((x - mu), 2) / (2 * np.pow(std, 2)))
    y_distribution = normal_distribution(x_distribution, mu, std)

    plt.grid(linestyle='--')
    plt.scatter(x, np.zeros(x.shape), marker='x', label='Datapoints')
    plt.plot(x_distribution, y_distribution, color='g', label='Normal-Distribution')
    plt.vlines(mu, ymin=0, ymax=np.max(y_distribution), linestyles='--', color='r', label=r'$\mu = $' + str(round(mu, 2)))
    plt.vlines(mu + std, ymin=0, ymax=normal_distribution(np.array([mu + std]), mu, std)[0], linestyles='--', color='orange', label=r'$\sigma = $' + str(round(std, 2)))
    plt.vlines(mu - std, ymin=0, ymax=normal_distribution(np.array([mu - std]), mu, std)[0], linestyles='--', color='orange')
    plt.legend()
    plt.title('Normal-Distribution Maximum Likelihood Estimation (MLE)')
    plt.show()
```

```python
plot_normal_distribution(np.array([-5, 7, 4, -3, -4, 6, 0.5, 2.5, 0.3, 4, 2, 0, 1, 2, -1]))
```
